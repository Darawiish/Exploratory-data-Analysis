### Step 3 - Review Types of Basic Univariate Plots

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

#### Fill in blanks with a brief description of each plot type:

  
Histogram:
- Histograms are used to analyze numerical data and presents the frequency of data values, good for showing skewness. 

Density Plot:
- A smoother version of a histogram, Kernel density is used to show the probability density of the graph which shows in a more smooth manner.

Box Plots:
- Also used to look at the spread of data and these visualizations give you more description than what you see in a typical histogram.

Pie Charts:
- Pie charts are good for visualizing the data, the value of each data as a fraction of a whole. 

Bar Chart:
- Bar charts have good data visualization of nominal/categorical data and solve the problem that arise with pie charts

Univariate scatter plot:
- Requires two parameters to plot against one another (x,y coordinates) and observes the relationship of the two.

Line Plot:
- Line plot is similar to a scatter plot with lines connecting the dots but rather ordered by some value such as the x-axis.

Strip Plot:
- Similar to a scatter plot but it is organized vertically and also into categories this adds for better visualization when comparing many x-variables.

Rug Plot:
- Good supporter plot similar to a histogram but rather with lines and the vertical height is ambiguous. Hard to read on its own.

Violin Plot:
- Similar to a box plot combined with a density plot and shows distribution of quantitative data across several categorical data points.

### Step 5 - Create Plots for Each Variable in Dataset

```python
# import seaborn
import seaborn as sns


# set standard plot type as seaborn plot
sns.set ()

# import matplotlib
import matplotlib.pyplot as plt 

# write command for plots to appear in jupyter cells automatically
%matplotlib inline
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

### Plot for 'age' column

```python
# sets figure size
plt.figure(figsize=(12,6))

# establishes figure with data and plot type
ax = sns.histplot(data=df, x='age', palette="viridis", bins=8)

# sets x and y labels
plt.xlabel("Age", fontsize=14)
plt.ylabel("Number of Managers", fontsize=14)

# sets title
plt.title("Distribution of Manager Ages in Survey", fontsize=15);
```

##### Insights:

[There is a positive correlation with age and manager, this data represents that managers are typically older]

### Plot for 'industry' column

```python
plt.figure(figsize=(12,6))

ax = sns.countplot(data=df, x='industry',palette="icefire")

plt.xticks(rotation=90)
plt.xlabel("Industry", fontsize=13)
plt.ylabel("Number of Managers in Industry", fontsize=13)
plt.title("Industries in Survey", fontsize=15);
```

##### Insights:

[The data here shows that managers are needed evenly in each industry equally]

### Plot for 'salary' column

```python
sns.displot(data = df, x = 'salary', kind = 'hist', bins=30)

plt.xlabel("Salary", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.title("Distribution of Salary Among Managers", fontsize=14);
```

```python
plt.figure(figsize=(12,6))

sns.violinplot(x=df['salary'])

plt.xlabel("Salary", fontsize=14)
plt.title("Density of Mana`ger Salaries");
```

##### Insights:

[The data shows evenly pretty much an evenly spread bell curve]

### Plot for 'industry_experience' column

```python
plt.figure(figsize=(12,6))

ax = sns.histplot(data=df, x='industry_experience', bins=20)

plt.xlabel("Industry Experience ", fontsize=13)
plt.ylabel("Number of Managers", fontsize=13)
plt.title("Industry Experience Among Managers", fontsize=16);
```

##### Insights:

[This data is not easily as interpretable as it looks, At first glance, the distribution looks normal and even. But its hard to make a correlation, the broad assumption would be that most (more than half) of the managers have an industry experience of 10-25 years.]

### Plot for 'education' column

```python
df['education'].value_counts()
```

```python
plt.figure(figsize=(10,10))

data = df['education'].value_counts()

labels = ["Bachelor's Degree", "Master's Degree", "Some College", "High School"]

colors = sns.color_palette('Accent')

plt.title("Manager's Education Levels", fontsize=16)

ax = plt.pie(data, colors = colors, labels = labels, autopct = '%.0f%%');
```

##### Insights:

[78% of managers have a Bachelors or Master's degree, and only few managers having less education we can conclude education and management to have a positive relationship. Such that your more likely to have a management job with
higher levels of education.]

## Part 2 - Summary Statistics

### Step 1 - Review Summary Statistics

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

- mean
> The average of a set of data, the sum of the whole divided by it count of the whole.
- median
> A midpoint in data organized by range, covers errors from the mean. 
- mode
> The most frequent occurring data point.
- central tendency
> The data value that is the center of a normal distribution plot.
- distribution
> It's the mathematical representation of a collection of data and how likely a certain outcome is. 
- variance
> How far each of the data points are from the mean, and from every other number. 
- standard deviation
> The square root of the variance, measure the dispersion of a data set.
